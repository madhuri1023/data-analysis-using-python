```python
# TASK to display dataframe and fill null values
import numpy as np
data={'animal':['cat','lion','snake','dog','lion','snake','lion','dog','cat','lion'],
      'age':[2.5,3,0.5,np.NaN,5,2,4.5,np.NaN,7,3],
      'visits':[1,3,2,3,2,3,1,1,2,1],
      'priority':['yes','yes','no','yes','no','no','no','yes','no','yes']}
data
```

```python
d=pd.DataFrame(data)
d
```

```python
#select only the rows where the number of visitsis greater than 3
d[d['visits']>3]
```

```python
#select the rows where the animalis a lion and age is less than 3
d[(d['animal']=='lion') & (d['age']>3)]
```

```python
#count the number of each type of animal in df 
d.isnull().sum()
```

```python
#finding the mean of age
d['age'].mean()
```

```python
#filling the null valueswith mean of age
d['age'].fillna(3.4375,inplace=True)
d
```

```python
#calculate the sum of all visits in df
d['visits'].sum()
```

```python
#count the number of each type ofanimal in df
d['animal'].value_counts()
```

```python
#change the yes or no values in priority column with true or false
d['priority']=d['priority'].replace({'yes':True,'no':False})
d
```

```python
#prints the animal names which are having maximum visits like a dataframe
d['visits'].max()
```

```python
d[d['visits']==d['visits'].max()][['animal']]
```

```python
#TASK design a two line plot
x=[1,2,3,4,5,6,7,8,9]
y=[18,16,14,12,10,8,6,4,2]
x1=np.arange(1,10)
y1=np.arange(1,10)
plt.plot(x,y,marker='o',color='red',mec='blue',ms=10,mfc='green',ls='dashed')
plt.plot(x1,y1,marker='D',color='green',mec='red',ms=10,mfc='blue')
plt.legend(['x&y line','x1&y1line'],loc='lower right') #lengend describes about the lines.
#plt.show()
plt.grid()
```
